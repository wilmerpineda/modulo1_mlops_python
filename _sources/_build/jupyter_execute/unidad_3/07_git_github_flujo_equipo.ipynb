# 🤝 Unidad 3: Git y GitHub: flujo de trabajo en equipo

En este notebook se simula un flujo de trabajo colaborativo usando **Git y GitHub** dentro de un proyecto de Machine Learning. El objetivo no es únicamente ejecutar comandos, sino comprender cómo se organiza el trabajo entre varias personas cuando se construye un sistema reproducible.


## Objetivos de aprendizaje

Al finalizar este notebook, el estudiante podrá:

- comprender la lógica de trabajo colaborativo con ramas y Pull Requests;
- identificar responsabilidades diferenciadas dentro de un equipo;
- reconocer por qué no se debe trabajar directamente sobre `main`;
- aplicar un flujo básico de Git y GitHub alineado con un proyecto de MLOps.


## 1. Escenario de trabajo

Imaginemos un equipo de tres personas que está construyendo un proyecto para predecir el **tiempo de permanencia de usuarios** en una plataforma digital.

El equipo está conformado por:

- **Ana**: administradora del repositorio
- **Bruno**: responsable de datos y pipeline
- **Camila**: responsable de modelado y experimentación

Cada integrante debe trabajar en su propia rama y contribuir al proyecto sin afectar directamente la rama principal.


## 2. ¿Por qué no trabajar en `main`?

La rama `main` debe representar una versión estable del proyecto. Si cada integrante modifica directamente esa rama, aparecen problemas como:

- pérdida de control sobre cambios;
- dificultad para revisar trabajo;
- alta probabilidad de conflictos;
- falta de trazabilidad del proceso.

Por ello, el flujo recomendado es:

1. crear una rama de trabajo;
2. desarrollar cambios;
3. hacer commits;
4. enviar la rama al remoto;
5. abrir un Pull Request;
6. revisar y aprobar antes del merge.


## 3. Flujo general del equipo

```{mermaid}
flowchart LR
    A[main estable] --> B[feature/data-pipeline]
    A --> C[feature/model-training]
    B --> D[Pull Request]
    C --> E[Pull Request]
    D --> F[Revisión]
    E --> F
    F --> G[Merge a main]
```

Este flujo permite que varias personas trabajen en paralelo sin romper la estabilidad del proyecto.


## 4. Paso 1: creación del repositorio

La persona administradora del repositorio inicia el proyecto y crea la estructura base.

### Flujo conceptual

1. crear carpeta del proyecto;
2. inicializar Git;
3. crear primer commit;
4. conectar con GitHub.


````python
# Estos comandos son ilustrativos y deben ejecutarse en terminal.

commands_repo_setup = [
    "git init",
    "git add .",
    'git commit -m "initial project structure"',
    "git branch -M main",
    "git remote add origin https://github.com/usuario/session-duration-mlops.git",
    "git push -u origin main",
]

for cmd in commands_repo_setup:
    print(cmd)
````

## 5. Paso 2: cada integrante crea su rama

Una vez existe la base del proyecto, cada integrante crea una rama descriptiva.

### Ejemplo de ramas

- Ana: `feature/repo-structure`
- Bruno: `feature/data-pipeline`
- Camila: `feature/model-training`


````python
team_branch_examples = {
    "Ana": "git checkout -b feature/repo-structure",
    "Bruno": "git checkout -b feature/data-pipeline",
    "Camila": "git checkout -b feature/model-training",
}

for member, command in team_branch_examples.items():
    print(f"{member}: {command}")
````

## 6. Paso 3: desarrollo de cambios en paralelo

Ahora cada persona realiza cambios específicos.

### Ana
- organiza carpetas;
- crea `.gitignore`;
- documenta estructura en `README.md`.

### Bruno
- desarrolla la generación de datos;
- prepara el pipeline base;
- propone archivos como `params.yaml` o `dvc.yaml`.

### Camila
- implementa el modelo base;
- calcula métricas;
- prepara integración con tracking.


## 7. Paso 4: commits pequeños y descriptivos

En un flujo profesional, los commits no deben ser ambiguos ni demasiado grandes.

### Buenos ejemplos

- `add initial folder structure`
- `add mock data generation script`
- `add baseline regression model`
- `update readme with execution steps`

### Mal ejemplo

- `cambios`

Un buen commit debe facilitar la revisión y la trazabilidad.


````python
good_commit_examples = [
    'git commit -m "add initial folder structure"',
    'git commit -m "add mock data generation script"',
    'git commit -m "add baseline regression model"',
]

for cmd in good_commit_examples:
    print(cmd)
````

## 8. Paso 5: envío de ramas al repositorio remoto

Una vez se completan cambios coherentes en una rama, estos deben enviarse a GitHub.


````python
push_examples = [
    "git push origin feature/repo-structure",
    "git push origin feature/data-pipeline",
    "git push origin feature/model-training",
]

for cmd in push_examples:
    print(cmd)
````

## 9. Paso 6: Pull Requests y revisión

Una vez la rama está en GitHub, no se hace merge inmediatamente. Primero se abre un **Pull Request**.

El Pull Request permite:

- revisar código;
- discutir cambios;
- validar que el trabajo sea coherente;
- decidir si la rama puede integrarse a `main`.

En nuestro escenario, Ana, como administradora del repositorio, revisa los Pull Requests de Bruno y Camila antes del merge.


## 10. Ejemplo de revisión colaborativa

Supongamos que Bruno abre un Pull Request para integrar el pipeline de datos. La revisión podría incluir preguntas como:

- ¿la estructura del archivo está alineada con el proyecto?
- ¿los nombres de variables son claros?
- ¿el código es reproducible?
- ¿la rama incluye cambios que no deberían estar ahí?

Este proceso es importante porque en MLOps no solo importa que el código funcione, sino que sea mantenible y comprensible.


## 11. Paso 7: integración controlada a `main`

Solo después de revisar y aprobar el Pull Request se realiza el merge a `main`.

```{mermaid}
flowchart LR
    A[feature/data-pipeline] --> B[Pull Request]
    B --> C[Review]
    C --> D[Approved]
    D --> E[Merge to main]
```

Así, `main` conserva su rol de rama estable.


## 12. Errores comunes en trabajo colaborativo

### ❌ Error 1. Trabajar directamente en `main`
Rompe el flujo de revisión y aumenta el riesgo de conflictos.

### ❌ Error 2. Un solo integrante hace todo el trabajo
El proyecto deja de ser colaborativo y no refleja la lógica del equipo.

### ❌ Error 3. Pull Requests vacíos o meramente formales
No agregan valor si no hay revisión real.

### ❌ Error 4. Commits gigantes y desordenados
Dificultan entender la evolución del proyecto.


## 13. Relación con la actividad integradora

Este flujo es exactamente el que se espera en la actividad del módulo.

Es decir, no basta con que exista un repositorio: debe haber evidencia de:

- ramas de trabajo;
- commits coherentes;
- Pull Requests;
- distribución de responsabilidades;
- integración organizada a `main`.

La actividad no evalúa solo el producto final, sino también el proceso de colaboración.
